```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

#1. An치lisis de facturaci칩n



## Tienda 1

```python
total1 = tienda['Precio'].sum()
print("Ingresos totales:", total1)
```

## Tienda 2

```python
total2 = tienda2['Precio'].sum()
print("Ingresos totales:",total2)
```

## Tienda 3

```python
total3 = tienda3['Precio'].sum()
print("Ingresos totales:",total3)
```

## Tienda 4

```python
total4 = tienda4['Precio'].sum()
print("Ingresos totales:",total4)
```

## Gr치fico

```python
import matplotlib.pyplot as plt
import numpy as np

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [total1, total2, total3, total4]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color=['orange', 'lime', 'blue', 'yellow'])
plt.title('Ingresos totales por tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos (MM)', color='black')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
ingresos_totales_todas_tiendas = total1 + total2 + total3 + total4
porcentaje_ingresos = [
    (total1 / ingresos_totales_todas_tiendas) * 100,
    (total2 / ingresos_totales_todas_tiendas) * 100,
    (total3 / ingresos_totales_todas_tiendas) * 100,
    (total4 / ingresos_totales_todas_tiendas) * 100
]

plt.figure(figsize=(8, 8))
plt.pie(porcentaje_ingresos, labels=tiendas, autopct='%1.1f%%', startangle=90, colors=['orange', 'lime', 'blue', 'yellow'])
plt.title('Ingresos Totales')
plt.show()
```
